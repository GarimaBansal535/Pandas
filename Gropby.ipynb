# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique

```python
len(df.groupby('Genre'))
# df['Genre'].nunique()
```

```python
# df.groupby('Genre').size()
df["Genre"].value_counts()
```

```python
genre=df.groupby('Genre')
genre.first()
genre.last()
genre.nth(6)
```

```python
genre.get_group('Drama')
```

```python
genre.groups  
```

```python
genre.describe()
```

```python
genre.sample(2,replace=True)
```

```python

genre.nunique()
```

```python
# agg method
# passing dict
genre.agg(
    {
        
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Meta_score':'min'
    }
)
```

```python
# passing list
genre.agg(['min','max','mean','sum'])
```

```python
genre.agg(
    {
        
        'IMDB_Rating':['mean','min'],
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Meta_score':'min'
    }
)
```
